# Matplotlib Barchart

```
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

1) Write a Python programming to display a bar chart of the popularity of Programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```

Languages=['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]


x = np.arange(len(Languages))

col= ['red','yellow','blue','green','m','black']
plt.bar(x,Popularity, label="Popularity",color=col)

plt.xticks(x,Languages)
plt.ylabel("Popularity")
plt.title('Programming languages')
plt.legend()

```

2) Write a Python program to create a bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```
women= [25, 32, 30, 35, 29]
men= [22, 30, 35, 35, 26]


x= np.arange(5)
plt.bar(x+0,women,color='r',width=.40, label='Women')
plt.bar(x+0.40,men,color='g',width=.40, label='Men')
plt.legend()
plt.show()
```

3) Write a Python program to add textures (black and white) to bars and wedges.
Note: Use bottom to stack the women's bars on top of the men's bars.

```

fig = plt.figure()
gender = fig.add_subplot(111)
gender.bar(x,women,color='white', edgecolor='black', hatch="//",label='women')
gender.bar(x,men,color='white', edgecolor='black', hatch='.',label='men')
gender.set_xticks([0,1,2,3,4,5])
plt.legend()

```

# Matplotlib Pie Chart

4) Write a Python programming to create a pie chart with a title of the popularity of Programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```
languages= ['Java', 'Python', 'PHP', 'JavaScript', 'C#','C++']
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]


color= ['r','b','y','m','g','black']

plt.pie(Popularity,labels= languages,colors=color,shadow=True)
plt.title("Popularity of Programming Languages")
plt.legend()
plt.show()
```
